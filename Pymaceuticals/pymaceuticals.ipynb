```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
#import warnings 
#warnings.filterwarnings("ignore")

# File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug data and Clinical Trial data
mouse_data_df = pd.read_csv(mouse_drug_data_to_load)
mouse_data_df.head()
#mouse_data_df.info()

clinical_data_df = pd.read_csv(clinical_trial_data_to_load)
clinical_data_df.head()
#clinical_data_df.info()
```

```python
# Merge the two dataframes using a left join
merge_table = pd.merge(clinical_data_df, mouse_data_df, on="Mouse ID", how="left")
merge_table.head()
# merge_table.info()
```

```python
### Tumor response to treatment ###
#
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
tumor_vol_mean_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Tumor Volume (mm3)': "mean"}
)

# Reset the index of the DataFrame and make 'Drug' and "Timepoint" as columns
tumor_vol_mean_df = tumor_vol_mean_df.reset_index(level=["Drug","Timepoint"])

# Preview DataFrame
tumor_vol_mean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_sem_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Tumor Volume (mm3)': lambda x: sem(x)}
)

# Reset the index of the DataFrame and make 'Drug' and "Timepoint" as columns
tumor_vol_sem_df = tumor_vol_sem_df.reset_index(level=["Drug","Timepoint"])

# Preview DataFrame
tumor_vol_sem_df
```

```python
# Minor Data Munging to Re-Format the DataFrames
# Create pivot tables for means and standard error with Timepoint as Index and Drug as Column
tumor_sem_pivot = tumor_vol_sem_df.pivot_table(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

tumor_mean_pivot = tumor_vol_mean_df.pivot_table(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
tumor_mean_pivot
```

```python
# Generate the Plot (with Error Bars) for 'Tumor response to Treatment' with 
# Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
# Create the data
drugs = []
drugs = list(tumor_mean_pivot.columns)

# Set x-axis and variables
x_axis = tumor_vol_mean_df['Timepoint'].unique()
capo_mean = tumor_mean_pivot[drugs[0]]
infu_mean = tumor_mean_pivot[drugs[2]]
keta_mean = tumor_mean_pivot[drugs[3]]
plac_mean = tumor_mean_pivot[drugs[5]]
capo_sem = tumor_sem_pivot[drugs[0]]
infu_sem = tumor_sem_pivot[drugs[2]]
keta_sem = tumor_sem_pivot[drugs[3]]
plac_sem = tumor_sem_pivot[drugs[5]]

# Draw horizontal lines with 0.25 transparency
#plt.hlines(y,xmin=0,xmax=1,alpha=0.25)

# Setting up the plot
# Plot 'Tumor response to Treatment' for Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
fig, ax = plt.subplots()

ax.errorbar(x_axis, capo_mean, capo_sem, fmt='o', color='red')
ax.errorbar(x_axis, infu_mean, infu_sem, fmt='^', color='blue')
ax.errorbar(x_axis, keta_mean, keta_sem, fmt='s', color='green')
ax.errorbar(x_axis, plac_mean, plac_sem, fmt='d', color='black')
capo_handle, = plt.plot(x_axis, capo_mean, color='red', label='Capomulin')
infu_handle, = plt.plot(x_axis, infu_mean, color='blue', label='Infubinol')
keta_handle, = plt.plot(x_axis, keta_mean, color='green', label='Ketapril')
plac_handle, = plt.plot(x_axis, plac_mean, color='black', label='Placebo')

# Set the plot title, lables and legend
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc=2)

# Save an image of the 'Tumor response to Treatment' chart so that we can view it in a folder
plt.savefig("Images/treatment.png")
plt.show()
```

```python
### Metastatic response to treatment ###
#
# Store the Mean Metastatic Site Data Grouped by Drug and Timepoint
meta_site_mean_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Metastatic Sites': "mean"}
)

# Preview DataFrame
meta_site_mean_df
```

```python
# Store the Standard Error of Metstatic Sites Grouped by Drug and Timepoint
meta_site_sem_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Metastatic Sites': lambda x: sem(x)}
)

# Preview DataFrame
meta_site_sem_df
```

```python
# Minor Data Munging to Re-Format the DataFrames
# Create pivot tables for means and standard error with Timepoint as Index and Drug as Column

# Reset the index of the DataFrames and make 'Drug' and "Timepoint" as columns
meta_site_mean_df = meta_site_mean_df.reset_index(level=["Drug","Timepoint"])
meta_site_sem_df = meta_site_sem_df.reset_index(level=["Drug","Timepoint"])

meta_site_sem_pivot = meta_site_sem_df.pivot_table(index='Timepoint', columns='Drug', values='Metastatic Sites')
#meta_site_sem_pivot

meta_site_mean_pivot = meta_site_mean_df.pivot_table(index='Timepoint', columns='Drug', values='Metastatic Sites')
meta_site_mean_pivot
```

```python
# Generate the Plot (with Error Bars) for 'Metastatic Response to Treatment' with 
# Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
# Create the data
meta_drugs = []
meta_drugs = list(meta_site_mean_pivot.columns)
```

```python
# Set x-axis and variables
meta_x_axis = meta_site_mean_df['Timepoint'].unique()
meta_capo_mean = meta_site_mean_pivot[meta_drugs[0]]
meta_infu_mean = meta_site_mean_pivot[meta_drugs[2]]
meta_keta_mean = meta_site_mean_pivot[meta_drugs[3]]
meta_plac_mean = meta_site_mean_pivot[meta_drugs[5]]
meta_capo_sem = meta_site_sem_pivot[meta_drugs[0]]
meta_infu_sem = meta_site_sem_pivot[meta_drugs[2]]
meta_keta_sem = meta_site_sem_pivot[meta_drugs[3]]
meta_plac_sem = meta_site_sem_pivot[meta_drugs[5]]

# Draw horizontal lines with 0.25 transparency
#plt.hlines(y,xmin=0,xmax=1,alpha=0.25)

# Setting up the plot
# Plot 'Metastatic Response to Treatment' with  Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
fig, ax = plt.subplots()

ax.errorbar(meta_x_axis, meta_capo_mean, meta_capo_sem, fmt='o', color='red')
ax.errorbar(meta_x_axis, meta_infu_mean, meta_infu_sem, fmt='^', color='blue')
ax.errorbar(meta_x_axis, meta_keta_mean, meta_keta_sem, fmt='s', color='green')
ax.errorbar(meta_x_axis, meta_plac_mean, meta_plac_sem, fmt='d', color='black')
meta_capo_handle, = plt.plot(meta_x_axis, meta_capo_mean, color='red', label='Capomulin')
meta_infu_handle, = plt.plot(meta_x_axis, meta_infu_mean, color='blue', label='Infubinol')
meta_keta_handle, = plt.plot(meta_x_axis, meta_keta_mean, color='green', label='Ketapril')
meta_plac_handle, = plt.plot(meta_x_axis, meta_plac_mean, color='black', label='Placebo')

# Set the plot title, lables and legend
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met Sites')
plt.legend(loc=2)

# Save an image of the 'Tumor response to Treatment' chart so that we can view it in a folder
plt.savefig("Images/spread.png")
plt.show()
```
