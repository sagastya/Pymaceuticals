```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
#import warnings 
#warnings.filterwarnings("ignore")

# File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug data and Clinical Trial data
mouse_data_df = pd.read_csv(mouse_drug_data_to_load)
mouse_data_df.head()
#mouse_data_df.info()

clinical_data_df = pd.read_csv(clinical_trial_data_to_load)
clinical_data_df.head()
#clinical_data_df.info()
```

```python
# Merge the two dataframes using a left join
merge_table = pd.merge(clinical_data_df, mouse_data_df, on="Mouse ID", how="left")
merge_table.head()
# merge_table.info()
```

```python
### Tumor response to treatment ###
#
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
tumor_vol_mean_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Tumor Volume (mm3)': "mean"}
)

# Reset the index of the DataFrame and make 'Drug' and "Timepoint" as columns
tumor_vol_mean_df = tumor_vol_mean_df.reset_index(level=["Drug","Timepoint"])

# Preview DataFrame
tumor_vol_mean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_sem_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Tumor Volume (mm3)': lambda x: sem(x)}
)

# Reset the index of the DataFrame and make 'Drug' and "Timepoint" as columns
tumor_vol_sem_df = tumor_vol_sem_df.reset_index(level=["Drug","Timepoint"])

# Preview DataFrame
tumor_vol_sem_df
```

```python
# Minor Data Munging to Re-Format the DataFrames
# Create pivot tables for means and standard error with Timepoint as Index and Drug as Column
tumor_sem_pivot = tumor_vol_sem_df.pivot_table(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

tumor_mean_pivot = tumor_vol_mean_df.pivot_table(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
tumor_mean_pivot
```

```python
# Generate the Plot (with Error Bars) for 'Tumor response to Treatment' with 
# Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
# Create the data
drugs = []
drugs = list(tumor_mean_pivot.columns)

# Set x-axis and variables
x_axis = tumor_vol_mean_df['Timepoint'].unique()
capo_mean = tumor_mean_pivot[drugs[0]]
infu_mean = tumor_mean_pivot[drugs[2]]
keta_mean = tumor_mean_pivot[drugs[3]]
plac_mean = tumor_mean_pivot[drugs[5]]
capo_sem = tumor_sem_pivot[drugs[0]]
infu_sem = tumor_sem_pivot[drugs[2]]
keta_sem = tumor_sem_pivot[drugs[3]]
plac_sem = tumor_sem_pivot[drugs[5]]

# Draw horizontal lines with 0.25 transparency
#plt.hlines(y,xmin=0,xmax=1,alpha=0.25)

# Setting up the plot
# Plot 'Tumor response to Treatment' for Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
fig, ax = plt.subplots()

ax.errorbar(x_axis, capo_mean, capo_sem, fmt='o', color='red')
ax.errorbar(x_axis, infu_mean, infu_sem, fmt='^', color='blue')
ax.errorbar(x_axis, keta_mean, keta_sem, fmt='s', color='green')
ax.errorbar(x_axis, plac_mean, plac_sem, fmt='d', color='black')
capo_handle, = plt.plot(x_axis, capo_mean, color='red', label='Capomulin', linewidth=0.5)
infu_handle, = plt.plot(x_axis, infu_mean, color='blue', label='Infubinol', linewidth=0.5)
keta_handle, = plt.plot(x_axis, keta_mean, color='green', label='Ketapril', linewidth=0.5)
plac_handle, = plt.plot(x_axis, plac_mean, color='black', label='Placebo', linewidth=0.5)

# Set the plot title, lables and legend
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc=2)
ax.grid()

# Save an image of the 'Tumor response to Treatment' chart so that we can view it in a folder
plt.savefig("Images/treatment.png")
plt.show()
```

```python
### Metastatic response to treatment ###
#
# Store the Mean Metastatic Site Data Grouped by Drug and Timepoint
meta_site_mean_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Metastatic Sites': "mean"}
)

# Preview DataFrame
meta_site_mean_df
```

```python
# Store the Standard Error of Metstatic Sites Grouped by Drug and Timepoint
meta_site_sem_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
    {'Metastatic Sites': lambda x: sem(x)}
)

# Preview DataFrame
meta_site_sem_df
```

```python
# Minor Data Munging to Re-Format the DataFrames
# Create pivot tables for means and standard error with Timepoint as Index and Drug as Column

# Reset the index of the DataFrames and make 'Drug' and "Timepoint" as columns
meta_site_mean_df = meta_site_mean_df.reset_index(level=["Drug","Timepoint"])
meta_site_sem_df = meta_site_sem_df.reset_index(level=["Drug","Timepoint"])

meta_site_sem_pivot = meta_site_sem_df.pivot_table(index='Timepoint', columns='Drug', values='Metastatic Sites')
#meta_site_sem_pivot

meta_site_mean_pivot = meta_site_mean_df.pivot_table(index='Timepoint', columns='Drug', values='Metastatic Sites')
meta_site_mean_pivot
```

```python
# Generate the Plot (with Error Bars) for 'Metastatic Response to Treatment' with 
# Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
# Create the data
meta_drugs = []
meta_drugs = list(meta_site_mean_pivot.columns)

# Set x-axis and variables
meta_x_axis = meta_site_mean_df['Timepoint'].unique()
meta_capo_mean = meta_site_mean_pivot[meta_drugs[0]]
meta_infu_mean = meta_site_mean_pivot[meta_drugs[2]]
meta_keta_mean = meta_site_mean_pivot[meta_drugs[3]]
meta_plac_mean = meta_site_mean_pivot[meta_drugs[5]]
meta_capo_sem = meta_site_sem_pivot[meta_drugs[0]]
meta_infu_sem = meta_site_sem_pivot[meta_drugs[2]]
meta_keta_sem = meta_site_sem_pivot[meta_drugs[3]]
meta_plac_sem = meta_site_sem_pivot[meta_drugs[5]]

# Draw horizontal lines with 0.25 transparency
#plt.hlines(y,xmin=0,xmax=1,alpha=0.25)

# Setting up the plot
# Plot 'Metastatic Response to Treatment' with  Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
fig, ax = plt.subplots()

ax.errorbar(meta_x_axis, meta_capo_mean, meta_capo_sem, fmt='o', color='red')
ax.errorbar(meta_x_axis, meta_infu_mean, meta_infu_sem, fmt='^', color='blue')
ax.errorbar(meta_x_axis, meta_keta_mean, meta_keta_sem, fmt='s', color='green')
ax.errorbar(meta_x_axis, meta_plac_mean, meta_plac_sem, fmt='d', color='black')
meta_capo_handle, = plt.plot(meta_x_axis, meta_capo_mean, color='red', label='Capomulin', linewidth=0.5)
meta_infu_handle, = plt.plot(meta_x_axis, meta_infu_mean, color='blue', label='Infubinol', linewidth=0.5)
meta_keta_handle, = plt.plot(meta_x_axis, meta_keta_mean, color='green', label='Ketapril', linewidth=0.5)
meta_plac_handle, = plt.plot(meta_x_axis, meta_plac_mean, color='black', label='Placebo', linewidth=0.5)

# Set the plot title, lables and legend
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met Sites')
plt.legend(loc=2)
ax.grid()

# Save an image of the 'Tumor response to Treatment' chart so that we can view it in a folder
plt.savefig("Images/spread.png")
plt.show()
```

```python
### Survival Rates ###
#
# Store the count of Mice grouped by Drug and Timepoint 
mouse_count_df = merge_table.groupby(
    ["Drug", "Timepoint"]    
).agg(
     mouse_count=pd.NamedAgg(column='Mouse ID', aggfunc="count")
)
mouse_count_df = mouse_count_df.rename(columns={
    "mouse_count":"Mouse Count"})

# Preview DataFrame
mouse_count_df
```

```python
# Minor Data Munging to Re-Format the DataFrame
# Reset the index of the DataFrame and make 'Drug' and "Timepoint" as columns
mouse_count_df = mouse_count_df.reset_index(level=["Drug","Timepoint"])
mouse_count_df
```

```python
# Create a pivot table with Timepoint as Index and Drug as Column
mouse_count_pivot = mouse_count_df.pivot_table(index='Timepoint', columns='Drug', values='Mouse Count')
mouse_count_pivot
```

```python
# Generate the Plot (accounting for percentages) of survival rates with 
# Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
# Create the data
surv_drugs = []
surv_drugs = list(mouse_count_pivot.columns)

# Set x-axis and variables
surv_x_axis = mouse_count_df['Timepoint'].unique()

# Calculate the initial count of mice for each drug trial with Drugs -'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
capo_mouse_total = mouse_count_pivot[surv_drugs[0]][0]
infu_mouse_total = mouse_count_pivot[surv_drugs[2]][0]
keta_mouse_total = mouse_count_pivot[surv_drugs[3]][0]
plac_mouse_total = mouse_count_pivot[surv_drugs[5]][0]

# Create the surival rates of mice at each Timepoint of the Drug
capo_mouse_pct = []
infu_mouse_pct = []
keta_mouse_pct = []
plac_mouse_pct = []
for x in mouse_count_pivot[surv_drugs[0]]:
    pct = round(x/capo_mouse_total * 100, 2)
    capo_mouse_pct.append(pct)
for x in mouse_count_pivot[surv_drugs[2]]:
    pct = round(x/infu_mouse_total * 100, 2)
    infu_mouse_pct.append(pct) 
for x in mouse_count_pivot[surv_drugs[3]]:
    pct = round(x/keta_mouse_total * 100, 2)
    keta_mouse_pct.append(pct)
for x in mouse_count_pivot[surv_drugs[5]]:
    pct = round(x/plac_mouse_total * 100, 2)
    plac_mouse_pct.append(pct)  

# Setting up the plot
# Plot 'Survival Rate percentages after Treatment' with  Drugs - 'Capomulin', 'Infubinol', 'Ketapril', 'Placebo'
surv_capo_handle, = plt.plot(surv_x_axis, capo_mouse_pct, marker='o', color='red', label='Capomulin', linewidth=0.5)
surv_infu_handle, = plt.plot(surv_x_axis, infu_mouse_pct, marker='^', color='blue', label='Infubinol', linewidth=0.5)
surv_keta_handle, = plt.plot(surv_x_axis, keta_mouse_pct, marker='s', color='green', label='Ketapril', linewidth=0.5)
surv_plac_handle, = plt.plot(surv_x_axis, plac_mouse_pct, marker='d', color='black', label='Placebo', linewidth=0.5)

# Set the plot title, lables and legend
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.legend(loc=3)
plt.grid()

# Save an image of the 'Tumor response to Treatment' chart so that we can view it in a folder
plt.savefig("Images/survival.png")
plt.show()
```

```python
### Summary Bar Graph ###
#
# Calculate the percent changes for each Drug                  

drugs = list(tumor_mean_pivot.columns)
drug_pct = []

capo_pct_change = (tumor_mean_pivot['Capomulin'][45] - tumor_mean_pivot['Capomulin'][0]) / 45 * 100
drug_pct.append(capo_pct_change)
ceft_pct_change = (tumor_mean_pivot['Ceftamin'][45] - tumor_mean_pivot['Ceftamin'][0]) / 45 * 100
drug_pct.append(ceft_pct_change)
infu_pct_change = (tumor_mean_pivot['Infubinol'][45] - tumor_mean_pivot['Infubinol'][0]) / 45 * 100
drug_pct.append(infu_pct_change)
keta_pct_change = (tumor_mean_pivot['Ketapril'][45] - tumor_mean_pivot['Ketapril'][0]) / 45 * 100
drug_pct.append(keta_pct_change)
naft_pct_change = (tumor_mean_pivot['Naftisol'][45] - tumor_mean_pivot['Naftisol'][0]) / 45 * 100
drug_pct.append(naft_pct_change)
plac_pct_change = (tumor_mean_pivot['Placebo'][45] - tumor_mean_pivot['Placebo'][0]) / 45 * 100
drug_pct.append(plac_pct_change)
prop_pct_change = (tumor_mean_pivot['Propriva'][45] - tumor_mean_pivot['Propriva'][0]) / 45 * 100
drug_pct.append(prop_pct_change)
rami_pct_change = (tumor_mean_pivot['Ramicane'][45] - tumor_mean_pivot['Ramicane'][0]) / 45 * 100
drug_pct.append(rami_pct_change)
stel_pct_change = (tumor_mean_pivot['Stelasyn'][45] - tumor_mean_pivot['Stelasyn'][0]) / 45 * 100
drug_pct.append(stel_pct_change)
zoni_pct_change = (tumor_mean_pivot['Zoniferol'][45] - tumor_mean_pivot['Zoniferol'][0]) / 45 * 100
drug_pct.append(zoni_pct_change)

drug_pct_change = dict(zip(drugs,drug_pct))
#drug_pct_change

drug_to_compare = {}
for key in ("Capomulin","Infubinol","Ketapril","Placebo"):
    drug_to_compare[key] = drug_pct_change[key]   
plot_data = pd.DataFrame()
plot_data['Drugs'] = list(drug_to_compare.keys())
plot_data['Pct'] = list(drug_to_compare.values())

# Configure plot and ticks
x_axis = np.arange(0, len(plot_data['Drugs']))
width = 0.98      # the width of the bars

fig, ax = plt.subplots()
rects = ax.bar(x_axis, plot_data['Pct'], width, color=(plot_data['Pct'] < 0).
                                        map({True: 'g', False: 'r'}), alpha=0.75, align="center")
# Set the plot title, lables, and axes
ax.set_title("Tumor Change Over 45 Day  Treatment")
ax.set_ylabel("% Tumor Volume Change")
ax.set_xticks(x_axis)
ax.set_xticklabels(plot_data['Drugs'])
#ax.grid()

# Use the function autolabel to label the bar graphs with the percent change for each drug
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)
        # Put the lable inside the column
        if p_height > 0.65: 
              label_position = height - (y_height * 0.65)
        elif p_height > 0.60:
              label_position = height - (y_height * 0.58)
        elif p_height > 0.50:
              label_position = height - (y_height * 0.52)
        else:
              label_position = height + (y_height * 0.15)
        
        ax.text(rect.get_x() + rect.get_width()/2.0, label_position,
                '%d' % int(height)+str('%'), color='white',
                ha='center', va='bottom')

autolabel(rects, ax)
ax.grid()

# Save an image of the 'Tumor Change Over 45 Day Treatment' chart so that we can view it in a folder
plt.savefig("Images/change.png")
plt.show()
     
```
